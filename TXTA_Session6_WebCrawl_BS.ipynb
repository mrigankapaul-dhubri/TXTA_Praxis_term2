```python
from bs4 import BeautifulSoup
import re
```

```python
url = "https://www.bankbazaar.com/reviews/hdfc-bank/credit-card.html?reviewPageNumber=1"
```

```python
import urllib.request
req = urllib.request.Request(
    url, 
    data=None, 
    headers={
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
    }
)

f = urllib.request.urlopen(req)
file = f.read().decode('utf-8')
```

```python
file
```

```python
doc = BeautifulSoup(file)
doc.head()
```

```python
containers = doc.find_all(name = 'li', attrs = {'class':"review-box"})
containers
```

```python
len(containers)
```

```python
reviews = []
for container in containers:
  review = container.find_all('div', attrs = {'class':'text_here review-desc-more'})[0]
  review_text = review.text.strip()
    
  reviews.append(review_text)
```

```python
reviews
```

```python
names = []
for container in containers:
  name = container.find_all('div', attrs={'class':'reviewer-profile'})[0]
  name_txt = name.text.strip()
  x = re.sub("\\n",'',name_txt)
  name_txt = re.sub("\s+"," ",x)
  names.append(name_txt)
```

```python
names[0] 
```

```python
x = re.sub("\\n",'',names[0])
y = re.sub("\s+"," ",x)
re.sub("[,]","",y)
```

```python
ratings = []
for container in containers:
  rating = container.find_all('div', attrs={"class":"rating-section review-user-score"})[0]
  rating = rating.text.strip()

  r = re.findall("(?<=\\n)[0-9.]+",rating)[0]   #backward extraction

  ratings.append(float(r))
  
```

```python
ratings
```

```python
df['Review_text'] = reviews
df['Rating'] = ratings
df['Other_info'] = names
```

```python
import time
```

```python
init = "https://www.bankbazaar.com/reviews/hdfc-bank/credit-card.html?reviewPageNumber="
```

```python
# df = pd.DataFrame(columns = ['Review_text', 'Rating','Other_info'])
doc_list = []

for page_number in range(1,3):
  url = init + str(page_number)

  req = urllib.request.Request(
      url, 
      data=None, 
      headers={
          'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
      }
      )

  f = urllib.request.urlopen(req)
  file = f.read().decode('utf-8')

  doc = BeautifulSoup(file)
  containers = doc.find_all(name = 'li', attrs = {'class':"review-box"})

  reviews = []
  names = []
  ratings = []
  for container in containers:
    review = container.find_all('div', attrs = {'class':'text_here review-desc-more'})[0]
    review_text = review.text.strip()  
    

    name = container.find_all('div', attrs={'class':'reviewer-profile'})[0]
    name_txt = name.text.strip()
    x = re.sub("\\n",'',name_txt)
    name_txt = re.sub("\s+"," ",x)
   
  
    rating = container.find_all('div', attrs={"class":"rating-section review-user-score"})[0]
    rating = rating.text.strip()
    r = re.findall("(?<=\\n)[0-9.]+",rating)[0]
    
    doc_list.append((review_text,float(r),x))

  time.sleep(3)


df = pd.DataFrame.from_records(doc_list, columns = ['Review_text', 'Rating','Other_info'] )

df.head()  
```
